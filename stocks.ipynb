```python
Stock.drop(['No. of Trades','Value Traded (SAR)','Volume Traded','% Change','Close','Industry Group','Symbol',"Change"], axis=1 ,inplace=True)
Stock
```

```python
# replace a column name
Stock.rename(columns={'Company Name': 'Stock_Date'}, inplace=True)
Stock.rename(columns={'Date': 'Stock_Open'}, inplace=True)
Stock.rename(columns={'Open': 'Stock_High'}, inplace=True)
Stock.rename(columns={'High': 'Stock_Low'}, inplace=True)
Stock.rename(columns={'Low': 'Stock_Close'}, inplace=True)
Stock.head()
```
